### Funkcje do wykorzystania

```python
def vizualize_test_result(pdf, statistic, rejection_region):
    """
    Parameters
    -------
    pdf: pandas.DataFrame
        Tablica zawierajca informacje niezbdne do wykrelenia funkcji gstoci prawdopodobiestwa wykorzystywanego w tecie rozkadu
        teoretycznego - index z wariantami zmiennej losowej i nadan nazw zmiennej (np. x) oraz pojedyncz kolumn z wartociami
        gstociami prawdopodobiestwa wyliczonymi dla poszczeg贸lnych wariant贸w o nadanej nazwie bazujcej na nazwie zmiennej (np. f(x)).
    statistic: float
        Warto statystyki testowej wyliczonej na podstawie pr贸by losowej.
    rejection_region: list
        Lista zawierajca trzy elementy pozwalajce na okrelenie obszaru krytycznego w nastpujcej kolejnoci:
         - warto dolnej granicy przedziau,
         - warto g贸rnej granicy przedziau,
         - "inner" lub "outer" w zale偶noci od tego, czy ma by zakolorowana wewntrzna, czy zewntrzna cz przedziau.
    """
    pdf_name= pdf.columns[0]
    
    fig, axes = plt.subplots(1, 1, figsize=(8, 3), facecolor='white')

    axes.plot(pdf.index, pdf[pdf_name], color="grey")

    if rejection_region[2]=="inner":
        axes.fill_between(pdf.index, pdf[pdf_name], 0, where=((pdf.index > rejection_region[0]) & (pdf.index<rejection_region[1])), color='red', alpha=0.5)
    elif rejection_region[2]=="outer":
        axes.fill_between(pdf.index, pdf[pdf_name], 0, where=((pdf.index < rejection_region[0]) | (pdf.index>rejection_region[1])), color='red', alpha=0.5)
    
    axes.vlines(x=statistic, ymin=0, ymax=np.max(pdf["pdf"]/3), color="blue")
    
    axes.set_xlabel(pdf.index.name)
    axes.set_ylabel(pdf_name)
```

### Zadanie 1: Weryfikacja hipotezy o wartoci przecitnej [7 pkt]

Przygotuj funkcj `mean_NHST()`, kt贸ra bdzie weryfikowa hipotez o wartoci przecitnej i wywietla wynik testu statystycznego zgodnie z wytycznymi zawartymi w doczonym do notatnika zestawie wzor贸w, w treci zadania oraz w docstring funkcji.

Poza zwracaniem zmiennych wymienionych w docsting funkcji, kt贸re s potrzebne do przeprowadzenia automatycznych test贸w funkcji powinna wykorzystywa udostpnion na pocztku notatnika funkcj `vizualize_test_result()` do generowania wykresu zawierajcego:
- funkcj gstoci prawdopodobiestwa rozkadu teoretycznego wykorzystywanego w tecie w zakresie od wartoci, dla kt贸rej dystrybuanta tego rozkadu przyjmuje warto 0.0001 do wartoci, dla kt贸rej dystrybuanta tego rozkadu przyjmuje warto 0.9999,
- zaznaczony wyr贸偶niajcym si kolorem fragmentem pola pod krzyw odpowiadajcy wyznaczonemu w tecie obszarowi krytycznemu,
- wyr贸偶niajc si kolorem pionow lini wskazujc obliczon warto statystyki.

```python
from scipy import stats
def mean_NHST(data, H0, H1, alpha):
    """
    Parameters
    -------
    data: DataFrame
        Tablica zawierajca domylny indeks i kolumn "X" wynikami pr贸by losowej.
    H0: float
        Warto przecitna przyjta jako hipoteza zerowa.
    H1: str
        Posta hipotezy alternatywnej, przyjmuje wartoci:
        - two-sided: warto przecitna jest r贸偶na od wartoci przyjtej w H0,
        - less: warto przecitna jest mniejsza od wartoci przyjtej w H0,
        - greater: warto przecitna jest wiksza od wartoci przyjtej w H0.
    alpha: float
        Warto okrelajca poziom istotnoci.
    
    Returns
    -------
    t: float
        Wyliczona na podstawie pr贸by losowej warto statystyki t.
    t_alpha: float
        Warto statystyki t_alpha.
    H: int
        Wynik testu statystycznego, przyjmuje warto:
        0 - gdy wynik testu istotnoci nie daje podstaw do odrzucenia H0 na rzecz H1 na poziomie istotnoci 1-alpha,
        1 - gdy nastpuje odrzucenie H0 na rzecz H1 na poziomie istotnoci 1-alpha.
    """
    # YOUR CODE HERE
    n = len(data["X"])
    #Sample mean and standard deviation
    mean = np.mean(data["X"])
    std = np.std(data["X"], ddof=1)
    #Standard error of the mean
    t = (mean - H0)/(std)
    t = t * (n**0.5)
    df = n - 1

    r_min = sp.stats.t.ppf(0.0001, df)   #t.ppf zwraca warto zmiennej losowej t (warto krytyczn), dla kt贸rej rozkad t-Studenta ma okrelon warto kwantyla 
    r_max = sp.stats.t.ppf(0.9999, df)
    r_variants = np.linspace(r_min,r_max,100)
    r1p = pd.DataFrame(data=stats.t.pdf(r_variants, df), index=r_variants, columns=["pdf"])  # t.pdf zwraca warto prawdopodobiestwa rozkadu w danym punkcie
    r1p.index.name = 'x'
    
    if H1 == 'two-sided':
        t_alpha = stats.t.ppf(1 - alpha/2, df)  # zwraca wart. krytyczn
        critical_region = abs(t) >= t_alpha
        vizualize_test_result(r1p, t, [-t_alpha, t_alpha, "outer"])
    elif H1 == 'less':
        t_alpha = stats.t.ppf(alpha, df)
        critical_region = t < t_alpha
        vizualize_test_result(r1p, t, [-np.inf, t_alpha, "inner"])
    elif H1 == 'greater':
        t_alpha = stats.t.ppf(1 - alpha, df)
        critical_region = t > t_alpha
        vizualize_test_result(r1p, t, [t_alpha, np.inf, "inner"])
    else:
        raise ValueError("Invalid H1 value. Choose from 'two-sided', 'less', or 'greater'.")

    H = 1 if critical_region else 0

    
    return t, t_alpha, H
```

```python
### Kom贸rka testowa
assert np.all(np.isclose(mean_NHST(test_data_1, 5.08, "two-sided", 0.05), (-1.9346677837081263, 2.1447866879169273, 0)))
assert np.all(np.isclose(mean_NHST(test_data_1, 5.08, "less", 0.05), (-1.9346677837081263, -1.7613101357748564, 1)))
assert np.all(np.isclose(mean_NHST(test_data_1, 5.08, "greater", 0.05), (-1.9346677837081263, 1.7613101357748562, 0)))

assert np.all(np.isclose(mean_NHST(test_data_1, 4.92, "two-sided", 0.1), (1.8994107469740968, 1.7613101357748562, 1)))
assert np.all(np.isclose(mean_NHST(test_data_1, 4.92, "less", 0.1), (1.8994107469740968, -1.3450303744546486, 0)))
assert np.all(np.isclose(mean_NHST(test_data_1, 4.92, "greater", 0.1), (1.8994107469740968, 1.345030374454649, 1)))
```

### Zadanie 2: Weryfikacja hipotezy o wariancji [8 pkt]

Przygotuj funkcj `variance_NHST()`, kt贸ra bdzie weryfikowa hipotez o wartoci wariancji i wywietla wynik testu statystycznego zgodnie z wytycznymi zawartymi w doczonym do notatnika zestawie wzor贸w, w treci zadania oraz w docstring funkcji.

Poza zwracaniem zmiennych wymienionych w docsting funkcji, kt贸re s potrzebne do przeprowadzenia automatycznych test贸w funkcji powinna wykorzystywa udostpnion na pocztku notatnika funkcj `vizualize_test_result()` do generowania wykresu zawierajcego:
- funkcj gstoci prawdopodobiestwa rozkadu teoretycznego wykorzystywanego w tecie w zakresie od wartoci, dla kt贸rej dystrybuanta tego rozkadu przyjmuje warto 0.0001 do wartoci, dla kt贸rej dystrybuanta tego rozkadu przyjmuje warto 0.9999,
- zaznaczony wyr贸偶niajcym si kolorem fragmentem pola pod krzyw odpowiadajcy wyznaczonemu w tecie obszarowi krytycznemu,
- wyr贸偶niajc si kolorem pionow lini wskazujc obliczon warto statystyki.

```python
def variance_NHST(data, H0, H1, alpha):
    """
    Parameters
    -------
    data: DataFrame
        Tablica zawierajca domylny indeks i kolumn "X" wynikami pr贸by losowej.
    H0: float
        Wariancji przyjta jako hipoteza zerowa.
    H1: str
        Posta hipotezy alternatywnej, przyjmuje wartoci:
        - two-sided: wariancja jest r贸偶na od wartoci przyjtej w H0,
        - less: wariancja jest mniejsza od wartoci przyjtej w H0,
        - greater: wariancja jest wiksza od wartoci przyjtej w H0.
    alpha: float
        Warto okrelajca poziom istotnoci.
    
    Returns
    -------
    chi2: float
        Wyliczona na podstawie pr贸by losowej warto statystyki chi2.
    chi2_alpha: float or list
        Warto statystyki chi2_alpha lub lista z dwoma wartociami chi2_alpha_1 i chi2_alpha_2
    H: int
        Wynik testu statystycznego, przyjmuje warto:
        0 - gdy wynik testu istotnoci nie daje podstaw do odrzucenia H0 na rzecz H1 na poziomie istotnoci 1-alpha,
        1 - gdy nastpuje odrzucenie H0 na rzecz H1 na poziomie istotnoci 1-alpha.
    """
    n = len(data["X"])
    std = np.std(data["X"])
    hi = n * (std**2)/(H0)
    df = n - 1
    r_min = sp.stats.chi2.ppf(0.0001, df)
    r_max = sp.stats.chi2.ppf(0.9999, df)
    r_variants = np.linspace(r_min,r_max,100)
    pdf = pd.DataFrame(data=stats.chi2.pdf(r_variants, df), index=r_variants, columns=["pdf"])  # t.pdf zwraca warto prawdopodobiestwa rozkadu w danym punkcie
    pdf.index.name = 'x'
    
    print(hi)

    if H1 == "two-sided":
        hi_alpha1 = sp.stats.chi2.ppf(alpha/2, n-1)
        hi_alpha2 = sp.stats.chi2.ppf(1 - alpha/2, n-1)
        print(hi_alpha1)
        result = hi <= hi_alpha1 or hi >= hi_alpha2
        H = 1 if result else 0 
        vizualize_test_result(pdf, hi, [hi_alpha1, hi_alpha2, "outer"])
        return hi, [hi_alpha1, hi_alpha2], H
    if H1 == "less":
        hi_alpha = sp.stats.chi2.ppf(alpha, n-1)
        result = hi<=hi_alpha
        vizualize_test_result(pdf, hi, [-np.inf, hi_alpha, "inner"])
    if H1 == "greater":
        hi_alpha = sp.stats.chi2.ppf(1-alpha, n-1)
        result = hi>=hi_alpha
        vizualize_test_result(pdf, hi, [hi_alpha, np.inf, "inner"])
    H = 1 if result else 0 

    return hi, hi_alpha, H
```

### Zadanie 3: Weryfikacja hipotezy o wska藕niku struktury (proporcji) [7 pkt]

Przygotuj funkcj `proportion_NHST()`, kt贸ra bdzie weryfikowa hipotez o wartoci wska藕nika struktury (proporcji) i wywietla wynik testu statystycznego zgodnie z wytycznymi zawartymi w doczonym do notatnika zestawie wzor贸w, w treci zadania oraz w docstring funkcji.

Poza zwracaniem zmiennych wymienionych w docsting funkcji, kt贸re s potrzebne do przeprowadzenia automatycznych test贸w funkcji powinna wykorzystywa udostpnion na pocztku notatnika funkcj `vizualize_test_result()` do generowania wykresu zawierajcego:
- funkcj gstoci prawdopodobiestwa rozkadu teoretycznego wykorzystywanego w tecie w zakresie od wartoci, dla kt贸rej dystrybuanta tego rozkadu przyjmuje warto 0.0001 do wartoci, dla kt贸rej dystrybuanta tego rozkadu przyjmuje warto 0.9999,
- zaznaczony wyr贸偶niajcym si kolorem fragmentem pola pod krzyw odpowiadajcy wyznaczonemu w tecie obszarowi krytycznemu,
- wyr贸偶niajc si kolorem pionow lini wskazujc obliczon warto statystyki.

```python
def proportion_NHST(data, H0, H1, alpha):
    """
    Parameters
    -------
    data: DataFrame
        Tablica zawierajca domylny indeks i kolumn "X" wynikami pr贸by losowej (True, False).
    H0: float
        Warto wska藕nika struktury przyjta jako hipoteza zerowa.
    H1: str
        Posta hipotezy alternatywnej, przyjmuje wartoci:
        - two-sided: warto wska藕nika struktury jest r贸偶na od wartoci przyjtej w H0,
        - less: warto wska藕nika struktury jest mniejsza od wartoci przyjtej w H0,
        - greater: warto wska藕nika struktury jest wiksza od wartoci przyjtej w H0.
    alpha: float
        Warto okrelajca poziom istotnoci.
    
    Returns
    -------
    u: float
        Wyliczona na podstawie pr贸by losowej warto statystyki u.
    u_alpha: float
        Warto statystyki u_alpha.
    H: int
        Wynik testu statystycznego, przyjmuje warto:
        0 - gdy wynik testu istotnoci nie daje podstaw do odrzucenia H0 na rzecz H1 na poziomie istotnoci 1-alpha,
        1 - gdy nastpuje odrzucenie H0 na rzecz H1 na poziomie istotnoci 1-alpha.
    """
    n = len(data["X"])
    affirmative_answers=0
    for x in data["X"]:
        if( x == True): 
            affirmative_answers= affirmative_answers + 1
    p = affirmative_answers / n
    u = p - H0
    temp = (H0*(1-H0)/n)**0.5
    u = u / temp
    r_min = sp.stats.norm.ppf(0.0001)
    r_max = sp.stats.norm.ppf(0.9999)
    r_variants = np.linspace(r_min, r_max, 100)
    pdf = pd.DataFrame( data = sp.stats.norm.pdf(r_variants), index = r_variants, columns=["pdf"])
    pdf.index.name = "x"
    if H1 == "two-sided":
        u_alpha = sp.stats.norm.ppf(1-alpha/2)
        result = abs(u) >= u_alpha
        vizualize_test_result(pdf, u , [-u_alpha, u_alpha, "outer"])
    if H1 == "less":
        u_alpha = sp.stats.norm.ppf(alpha)
        result = u <= u_alpha
        vizualize_test_result(pdf, u , [-np.inf, u_alpha, "inner"])
    if H1 == "greater":
        u_alpha = sp.stats.norm.ppf(1-alpha)
        result = u >= u_alpha
        vizualize_test_result(pdf, u , [u_alpha, np.inf, "inner"])
        
    H = 1 if result else 0 

    return u, u_alpha, H
```

Korzystajc z funkcji `scipy.stats.ttest_1samp()` zweryfikuj hipotez, 偶e rednia waga element贸w schodzcych z pierwszej linii produkcyjnej jest r贸wna 0.5kg wzgldem hipotezy alternatywnej, 偶e rednia waga element贸w schodzcych z pierwszej linii produkcyjnej jest wy偶sza od 0.5kg. W obliczeniach przyjmij poziom istotnoci 1 - alpha = 0.9.

Zapisz obliczon warto pvalue do zmiennej `pvalue_2`, oraz przygotuj zmienn `H_2` zawierajc wynik testu statystycznego (0 - gdy wynik testu istotnoci nie daje podstaw do odrzucenia H0 na rzecz H1 na poziomie istotnoci 1-alpha, 1 - gdy nastpuje odrzucenie H0 na rzecz H1 na poziomie istotnoci 1-alpha).

```python
alpha = 0.1
pvalue_2 = sp.stats.ttest_1samp(weights_1, 500, alternative= "greater").pvalue
print(pvalue_2)

if(pvalue_2 <= alpha):
    H_2 = 1
elif(pvalue_2 >= (1 - alpha)):
    H_2=1
else:
    H_2=0
```

```python
### Kom贸rka testowa
assert np.isclose(pvalue_2, 0.13688813254569193)
assert H_2 == 0
```
